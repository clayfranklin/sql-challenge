----

#### Latitude and Temperature
By observing temperatures based on their geopositional cordinates, it is apparent that the temperature is on average higher when approaching the equator, and lower as approaching the poles.
#### Latitude and Humidity
The humidity rate near the equator rarley dropped below 60%.  
#### Latitude and Wind Speed
The wind speed reach their highest speeds along the poles and the equator.  This is consistent with general hemisphere circular air patterns. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

print("Beggining Data Retrieval")
print("------------------------------------")

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="

name_of_city = []
cloudiness = []
temp = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []


# Loop through the list of cities and perform a request for data on each
    
for city in cities:
    response = requests.get(query_url + city).json()
        #clouds.append(response['clouds'])
    if response["cod"] != "404":
        name_of_city.append(response['name'])
        cloudiness.append(response['clouds']['all'])
        temp.append(response['main']['temp'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        wind_speed.append(response['wind']['speed'])
        city_for_log= response["name"]
        print(f"Processing Record | {city_for_log}")
        
    else:
        print("City not Found")
    time.sleep = (1.03)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
cities_df2 = pd.DataFrame({"City" : name_of_city,
                   "Cloudiness" : cloudiness,
                   "Country" : country,
                   "Date" : date,
                   "Humidity" : humidity,
                   "Lat"  : lat,
                   "Lng" : lng,
                   "Max Temp" : max_temp,
                   "Wind": wind_speed,
                  })
cities_df2.count()
```

```python
cities_df2.to_csv(output_data_file, index=False)
```
